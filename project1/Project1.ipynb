```python
path = 'data\Avengers_Endgame.pdf'

from tika import parser
import numpy as np

raw = parser.from_file(path)
temp_scenes = raw['content'].split('INT.')
separator = 'EXT.'
scenes = np.array([s for s in temp_scenes if separator not in s ])
for scene in temp_scenes :
    if separator in scene:
        scenes = np.append(scenes, scene.split(separator))
noices = [',', '!', ':', 'FRIDAY', 'F O R', 'END', '?']

actors_per_scene = []
for scene in scenes:
        tmp = [s.split('(')[0].strip() for s in set(scene.split('\n')) if s.isupper() and not (len([item for item in noices if item in s]) > 0) and len(s) < 20 and s]
        actors_per_scene.append(tmp)

new_actors_per_scene = []
for lst in actors_per_scene:
    if lst:
        tmp = []
        for act in lst:
            if act != '' and act not in tmp:
                tmp.append(act)
        new_actors_per_scene.append(tmp)
```
